```
from tensorflow import keras
import tensorflow as tf
from tqdm import tqdm
```

```
from tensorflow.python.keras import backend as K
config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth = True  # dynamically grow the memory used on the GPU
config.log_device_placement = True  # to log device placement (on which device the operation ran)
sess = tf.compat.v1.Session(config=config)
K.set_session(sess)

tf.test.is_built_with_cuda()
```

```
from tensorflow.keras.applications.resnet50 import ResNet50
```

```
model = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3),pooling='avg')
```

```
model.summary()
```

```
from tensorflow.keras.preprocessing import image
```

```
# with open('../data/Flickr_Data/Flickr_Data/Flickr_TextData/Flickr_8k.trainImages.txt') as f: 
#     trainer = f.readlines()
with open('../data/Flickr_Data/Flickr_Data/Flickr_TextData/Flickr_8k.devImages.txt') as f:
    validator = f.readlines()
with open('../data/Flickr_Data/Flickr_Data/Flickr_TextData/Flickr_8k.testImages.txt') as f: 
    tester = f.readlines()
```

```
# train = [s[:-1] for s in trainer]
val = [s[:-1] for s in validator]
test = [s[:-1] for s in tester]
```

```
len(val)
len(test)
```

```
path = '../data/Flickr_Data/Flickr_Data/Images/'

# imgDataTrain = {}
# for ix in tqdm(range(len(train))):
#     imPath = path+train[ix]
#     img = image.load_img(imPath,target_size=(224,224))
#     img_array = image.img_to_array(img)
#     img_array = img_array.reshape((1,224,224,3))
#     out = model.predict(img_array)
#     imgDataTrain[train[ix]] = out

imgDataVal = {}
for ix in tqdm(range(len(val))):
    imPath = path+val[ix]
    img = image.load_img(imPath,target_size=(224,224))
    img_array = image.img_to_array(img)
    img_array = img_array.reshape((1,224,224,3))
    out = model.predict(img_array)
    imgDataVal[val[ix]] = out 

imgDataTest = {}
for ix in tqdm(range(len(test))):
    imPath = path+test[ix]
    img = image.load_img(imPath,target_size=(224,224))
    img_array = image.img_to_array(img)
    img_array = img_array.reshape((1,224,224,3))
    out = model.predict(img_array)
    imgDataTest[test[ix]] = out  
```
